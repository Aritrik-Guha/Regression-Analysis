```python
x=df['sex']
for i in range(len(df)):
    if(x[i]=='female'):
        x[i]=0
    else:
        x[i]=1
```

```python
y=df['smoker']
for i in range(len(df)):
    if(y[i]=='yes'):
        y[i]=1
    else:
        y[i]=0
```

```python
x=df['region']
for i in range(len(df)):
    if(x[i]=='southeast'):
        x[i]=1
    elif(x[i]=='southwest'):
        x[i]=2
    elif(x[i]=='northeast'):
        x[i]=3
    else:
        x[i]=4
```

```python
X=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,6].values
y=y.reshape(-1,1)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```
